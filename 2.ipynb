```python
data=pd.read_csv(r"D:/Data Set/train.csv")
```

```python
pd.set_option("display.max_rows",None)
pd.set_option("display.max_columns",None)
```

```python
percent=data.isnull().sum()/data.shape[0]*100
```

```python
dropc=percent[percent>20].keys()
dropc
```

```python
data2=data.drop(columns=dropc)
data2.shape
```

```python
data2.isnull().sum()
```

```python
numericvar=data2.select_dtypes(include=["int64","float64"])

numericvar.shape
```

```python
numericvar.isnull().sum()
```

```python
missing_num_var=[ var for var in numericvar.columns if numericvar[var].isnull().sum() > 0]
missing_num_var
```

```python
# to check the distribution
```

```python
plt.figure(figsize=(20,20))
for i ,var in enumerate(missing_num_var):
    plt.subplot(2,2,i+1)
    sns.distplot(numericvar[var],bins=20,kde_kws={"linewidth": 5,"color":"red"})
    
```

```python
numericvar[missing_num_var].mean()
```

```python
datamean=numericvar.fillna(numericvar.mean())
```

```python
datamediann=numericvar.fillna(numericvar.median())
```

```python

plt.figure(figsize=(20,20))
for i ,var in enumerate(missing_num_var):
    plt.subplot(2,2,i+1)
    sns.distplot(numericvar[var],bins=20,kde_kws={"linewidth": 2,"color":"black"})
    sns.distplot(datamean[var],bins=20,kde_kws={"linewidth": 2,"color":"green"})
    sns.distplot(datamediann[var],bins=20,kde_kws={"linewidth": 2,"color":"red"})
```

```python
#to view data
concat=pd.concat([
    numericvar[missing_num_var],
    datamean[missing_num_var],
    datamediann[missing_num_var]],axis=1
               )
concat
```
